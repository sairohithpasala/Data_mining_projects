```python
# Import statements
#Array processing
import numpy as np
import matplotlib.pyplot as plt
#Data analysis, wrangling and common exploratory operations
import pandas as pd
from pandas import Series, DataFrame
import sklearn.metrics as metrics
from sklearn.metrics import classification_report 
```

```python
#read the csv file into a Pandas data frame
df = pd.read_csv('cardio_train.csv', sep=';')

#return the first 5 rows of the dataset
df.head()
```

```python
#Syntax to print last 5 rows of the dataset
df.tail()
```

```python
#3) Print the class labels.
df['cardio'].tolist()
```

```python
#Target vriable visulisation
import seaborn as sns
sns.set_theme(style="whitegrid")
cardio_data=df.replace([0,1],["Not Exists","Exists"])
#seaborn countplot graph syntax for target variable
ax = sns.countplot(x="cardio", palette="Set3",data=cardio_data)
```

```python
 #split dataset into features and target variable
# ignore "id" as it doesnt co-relate with other features and is unique for all rows
# dropping cardio as it is our target variable
X = df.drop(['id','cardio',],axis=1) # Features
y = df.cardio # Target variable
#column_names=['id','age','gender','height','weight','ap_hi','ap_lo','cholesterol','gluc','smoke','alco','active']
#print (X)
#print (y)
```

```python
# https://stackoverflow.com/questions/39409866/correlation-heatmap
import seaborn as sns

Var_Corr = df.corr()
# plot the heatmap and annotation on it
sns.heatmap(Var_Corr, xticklabels=Var_Corr.columns, yticklabels=Var_Corr.columns)
```

```python
#Splitting and Training data set : 
from sklearn import tree
from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1) # 70% training and 30% test

#Using Gini Index
model = tree.DecisionTreeClassifier( random_state=2,max_depth=3,min_samples_leaf=5)# Syntax for Decision Tree Classifier
model.fit(X_train, y_train) #fitting our training data into classifier 
y_predict = model.predict(X_test)#syntax to predict the target variable

#Syntax to print Accuracy and classification report
from sklearn.metrics import accuracy_score
print("The Accuracy is:", accuracy_score(y_test,y_predict)*100,"\n")
print("Classification Report: \n",classification_report(y_test,y_predict))
```

```python
from sklearn.metrics import confusion_matrix
#Syntax to print confusion matrix using Gini Index
pd.DataFrame(
    confusion_matrix(y_test, y_predict),
    columns=['Predicted Not Cardio', 'Predicted Cardio'],
    index=['True Not Cardio', 'True Cardio']
)
```

```python
#plotting the Confusion matrix in 2D using Gini Index
from sklearn.metrics import confusion_matrix
from sklearn import metrics

print("Confusion Matrix")
metrics.plot_confusion_matrix(model, X_test, y_test,cmap=plt.cm.Blues,display_labels=["Not Cardio","Cardio"])

plt.show()
```

```python
#decision tree visualization using GIni Index
from sklearn import tree
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(25, 8))
tree.plot_tree(model, fontsize=12, feature_names = df.columns)
plt.show()
```

```python
# using Entropy

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1) # 70% training and 30% test
model = tree.DecisionTreeClassifier( random_state=2,max_depth=3,min_samples_leaf=5)# Syntax for Decision Tree Classifier
model = tree.DecisionTreeClassifier( criterion= 'entropy' , random_state=100,max_depth=3,min_samples_leaf=5)
model.fit(X_train, y_train)#fitting our training data into classifier
y_predict = model.predict(X_test)#syntax to predict the target variable

#Syntax to print Accuracy and classification report
from sklearn.metrics import accuracy_score
print("The Accuracy is:", accuracy_score(y_test,y_predict)*100,"\n")
print("Classification Report: \n",classification_report(y_test,y_predict))
```

```python
#Confusion matrix using Entropy
pd.DataFrame(
    confusion_matrix(y_test, y_predict),
    columns=['Predicted Not Cardio', 'Predicted Cardio'],
    index=['True Not Cardio', 'True Cardio']
)
```

```python
#plotting the Confusion matrix in 2D using Entropy
from sklearn.metrics import confusion_matrix
from sklearn import metrics

print("Confusion Matrix")
metrics.plot_confusion_matrix(model, X_test, y_test,cmap=plt.cm.Greens,display_labels=["Not Cardio","Cardio"])
plt.show()
```

```python
# plot Decision Tree using Entropy
fig, ax = plt.subplots(figsize=(25, 8))
tree.plot_tree(model, fontsize=12, feature_names = df.columns)
plt.show()
```

```python
# Naive Bayes Classifier
from sklearn.naive_bayes import GaussianNB
#using default package attribute/fuction GaussianNB to implement Naive Bayes Classifier

gnb = GaussianNB(var_smoothing= 12)#var_smoothing essentially widens (or "smooths") the curve and accounts for more samples that are further away from the distribution
y_pred = gnb.fit(X_train, y_train).predict(X_test)#fitting the Gaussian with testing data to predict the target variable
print("Number of mislabeled points out of a total %d points : %d"  % (X_test.shape[0], (y_test != y_pred).sum()))
print("The Accuracy is:", accuracy_score(y_test,y_pred)*100,"\n")
print("Classification Report: \n",classification_report(y_test,y_pred))
```
